```python
from fastai.text import *
from fastai.text.learner import load_learner
import pandas as pd
```

```python
learn = load_learner('models')
```

```python
data = pd.read_csv('tweet.csv')
tweet_list = data['0'].values.tolist()
learn.data.add_test(tweet_list)
```

```python
preds,y = learn.get_preds(ds_type=DatasetType.Test, ordered=True)
```

```python
c2i = learn.data.c2i
i2c = {v:k for k,v in c2i.items()}
```

```python
prediction = preds.argmax(axis=1).tolist()
pred_label = [i2c[i] for i in prediction]
```

```python
pred_label
```

```python
data['pred'] = pd.Series(pred_label)
data.to_csv('tweet_with_result.csv')
```
